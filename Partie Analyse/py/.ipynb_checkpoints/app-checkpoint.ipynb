**Nous importons les bibliothèques Pandas et Matplotlib, ainsi que le fichier 'connexion_BD' qui permet de se connecter à la base de données.**

```python
import pandas as pd
import matplotlib.pyplot as plt
from connexion_BD import connectToDb

mydb = connectToDb()
my_cursor = mydb.cursor()
```

```python
my_cursor.execute("SELECT p.Age, COUNT(ap.idAliment) as nb_aliments FROM alimentprefdej ap JOIN aliment a ON ap.idAliment = a.idAliment JOIN personne p ON ap.idPersonne = p.idPersonne GROUP BY p.Age")

resultats = my_cursor.fetchall()

Age = [r[0] for r in resultats]
nb_aliments = [r[1] for r in resultats]


df = pd.read_sql_query(my_cursor.statement, mydb)


plt.bar(df['Age'], df['nb_aliments'], color='green')
plt.xlabel("Âge", fontsize=10, fontweight='bold')
plt.ylabel("Nombre d'aliments consommés", fontsize=10, fontweight='bold')
plt.title("Nombre d'aliments consommés par âge pour le petit-déjeuner", fontsize=14, fontweight='bold')
plt.show()
```

```python
my_cursor.execute("SELECT p.Age, COUNT(ap.idAliment) as nb_aliments FROM alimentprefdiner ap JOIN aliment a ON ap.idAliment = a.idAliment JOIN personne p ON ap.idPersonne = p.idPersonne GROUP BY p.Age")

resultats = my_cursor.fetchall()

Age = [r[0] for r in resultats]
nb_aliments = [r[1] for r in resultats]


df = pd.read_sql_query(my_cursor.statement, mydb)


plt.bar(df['Age'], df['nb_aliments'], color='green')
plt.xlabel("Âge", fontsize=10, fontweight='bold')
plt.ylabel("Nombre d'aliments consommés", fontsize=10, fontweight='bold')
plt.title("Nombre d'aliments consommés par âge pour le dîner", fontsize=14, fontweight='bold')
plt.show()
```

```python
my_cursor.execute("SELECT p.Civilite, COUNT(ap.idAliment) AS nb_aliments FROM personne p INNER JOIN alimentprefdej ap ON p.idPersonne = ap.idPersonne GROUP BY p.Civilite, p.idPersonne")

resultats = my_cursor.fetchall()

Civilite = [r[0] for r in resultats]
nb_aliments = [r[1] for r in resultats]

# Récupérer les données et les stocker dans un DataFrame pandas
df = pd.read_sql_query(my_cursor.statement, mydb)

plt.boxplot([df.loc[df['Civilite'] == 'Mme']['nb_aliments'], df.loc[df['Civilite'] == 'M']['nb_aliments']],
            labels=['Femme', 'Homme'])
plt.title('Nombre moyen d\'aliments consommés au petit-déjeuner')
plt.ylabel('Nombre d\'aliments')
plt.legend(['Femme', 'Homme'])
plt.show()
```

```python
my_cursor.execute("SELECT personne.idPersonne, aliment.`Energie, Règlement UE N° 1169/2011 (kcal/100 g)` as valeur_energetique FROM personne LEFT JOIN alimentprefdej ON personne.idPersonne = alimentprefdej.idPersonne LEFT JOIN aliment ON alimentprefdej.idAliment = aliment.idAliment UNION ALL SELECT personne.idPersonne, aliment.`Energie, Règlement UE N° 1169/2011 (kcal/100 g)` as valeur_energetique FROM personne LEFT JOIN alimentprefdiner ON personne.idPersonne = alimentprefdiner.idPersonne LEFT JOIN aliment ON alimentprefdiner.idAliment = aliment.idAliment;")

resultats = my_cursor.fetchall()

valeur_energetique = [r[1] for r in resultats]

# Récupérer les données et les stocker dans un DataFrame pandas
df = pd.read_sql_query(my_cursor.statement, mydb)


df['valeur_energetique'] = pd.to_numeric(df['valeur_energetique'], errors='coerce')
df = df.dropna(subset=['valeur_energetique'])
# Création de la boîte à moustache
plt.boxplot(df['valeur_energetique'].astype(float))
plt.title('Répartition des valeurs énergétiques des aliments consommés')
plt.ylabel('Energie (kcal/100 g)')
plt.show()
```

```python
my_cursor.execute("SELECT a.NomAliment, COUNT(*) AS NbPersonnes FROM aliment a JOIN alimentprefdej ad ON a.idAliment = ad.idAliment GROUP BY a.NomAliment ORDER BY NbPersonnes DESC LIMIT 5")
resultats = my_cursor.fetchall()

NomAliment = [r[0] for r in resultats]
NbPersonnes = [r[1] for r in resultats]
# Récupérer les données et les stocker dans un DataFrame pandas
df = pd.read_sql_query(my_cursor.statement, mydb)

plt.bar(df['NomAliment'], df['NbPersonnes'])
plt.xticks(rotation=90)
plt.xlabel('Aliment')
plt.ylabel('Nombre de personnes')
plt.title('Les aliments les plus consommés au petit-déjeuner')
plt.show() 
```
